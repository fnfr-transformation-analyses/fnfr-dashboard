```python
import pandas as pd
import re

data = pd.read_csv('data/affiliationsChercheurs.csv', sep=';')
data
```

```python
def typeEtablissement(etablissement: str)-> str:
    text = etablissement.lower()
    if re.match(".*universit*", text):
        return 'Académique'
    if re.match(".*research institute*", text):
        return 'Académique'
    
    if re.match(".*inc\..*", text):
        return 'Industriel'

    if re.match(".*LLC.*", etablissement):
        return 'Industriel'
    

data['type établissement'] = data['établissement'].apply(lambda x: typeEtablissement(x))
data[data['type établissement'].isnull()]
```
