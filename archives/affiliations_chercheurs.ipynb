```python
import pandas as pd

data = pd.read_excel('fnrf_transformation_chercheur-e-s.xlsx')
data = data[[
    'Titre de la demande / Application Title',
    'Chercheur-e principal-e désigné-e / Nominated Principal Investigator (PI)',
    'Affiliation chercheur-e principal-e / Principal Investigator Affiliation',
    'Cochercheur-e-s  / Co-principal investigators (CPIs)',
    'Co-candidats / \nCo-applicants',
    'Collaborateurs / Collaborators',
]]

data = data.to_dict('records')

chercheursProjets = []

for r in data:
    chercheur_principal = r['Chercheur-e principal-e désigné-e / Nominated Principal Investigator (PI)']

    chercheursProjets.append(
        {
            'chercheur': chercheur_principal.strip(),
            'projet': r['Titre de la demande / Application Title'],
            # 'rôle': 'Chercheur principal désigné',
            # 'établissement': r['Affiliation chercheur-e principal-e / Principal Investigator Affiliation']
        }

    )

    cochercheurs = r['Cochercheur-e-s  / Co-principal investigators (CPIs)'].split(";")
    cocandidats =  r['Co-candidats / \nCo-applicants'].split(";")
    collaborateurs = r['Collaborateurs / Collaborators'].split(";")
   
    for cochercheur in cochercheurs:
        chercheursProjets.append(
            {
                'chercheur': cochercheur.strip(),
                'projet': r['Titre de la demande / Application Title'],
            })
    
    for cocandidat in cocandidats:
        chercheursProjets.append(
            {
            'chercheur': cocandidat.strip(),
            'projet': r['Titre de la demande / Application Title'],
        })

    for collaborateur in collaborateurs:
        chercheursProjets.append(
            {
            'chercheur': collaborateur.strip(),
            'projet': r['Titre de la demande / Application Title'],
        })

chercheursProjets = pd.DataFrame(chercheursProjets)
chercheursProjets = chercheursProjets.drop_duplicates()
chercheursProjets.to_csv('chercheurs_projets.csv', index=False)
chercheursProjets
```

```python
data = pd.read_csv('affiliations_chercheurs.tsv', sep='\t')
data = data.drop_duplicates(subset='chercheur')

affiliationsChercheurs = data.merge(chercheursProjets, on='chercheur',  how='outer')
affiliationsChercheurs.drop_duplicates(subset='chercheur')
affiliationsChercheurs.to_csv('affiliationsChercheurs.csv', index=False)
affiliationsChercheurs.sort_values(by='chercheur')
```
